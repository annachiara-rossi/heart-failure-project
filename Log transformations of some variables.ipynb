```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(130298)
```

```python
data = pd.read_csv('train_data_drugs.csv')
data.set_index('inpatient.number', inplace = True)
data.head()
```

```python
var_to_log = ['creatinine.enzymatic.method', 'urea', 'glomerular.filtration.rate', 
              'cystatin', 'lymphocyte.count', 'neutrophil.count',
              'activated.partial.thromboplastin.time', 'prothrombin.time.ratio',
              'glutamyltranspeptidase','indirect.bilirubin','alkaline.phosphatase',
              'globulin','direct.bilirubin','cholesterol','low.density.lipoprotein.cholesterol','triglyceride']
```

```python
len(var_to_log)
```

```python
fig, axes = plt.subplots(2,8,sharey=True, figsize=(21,12))
fig.text(-0.01, 0.5, '', va='center', rotation='vertical', size = 15)
fig.suptitle('', size = 22)

i = 0
j = 0
for column_name in var_to_log[:8]*2:
    
    if i==0:
        sns.histplot(ax=axes[i,j], x = data[var_to_log[j]], kde = True, stat= 'probability')
    if i==1:
        sns.histplot(ax=axes[i,j], x = np.log(data[var_to_log[j]]), kde = True, stat= 'probability')
    
    axes[i,j].set_xlabel(column_name, size = 15)
    axes[i,j].set_ylabel('')
    j += 1
    if j == 8:
        j = 0
        i += 1
        
fig.tight_layout()
plt.show()
```
